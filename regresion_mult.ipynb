```python
import pandas as pd 
import numpy as np
```

```python
df = pd.read_excel("C:/Users/oscar/Downloads/Ingresos 1.xlsx")
df.head(3)
```

```python
umbral = df['Gasto'].median() 
df['Gasto_Alto'] = (df['Gasto'] > umbral).astype(int)
```

```python
df
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Variables para los ejes
x = df['Ingreso']  # Eje X
y = df['Edad']     # Eje Y
z = df['Gasto']    # Eje Z

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Graficar los puntos en 3D
ax.scatter(x, y, z, c='r', marker='o')

# Etiquetas de los ejes
ax.set_xlabel('Ingreso')
ax.set_ylabel('Edad')
ax.set_zlabel('Gasto')

plt.show()
```

```python
import statsmodels.api as sm
#  estimación y validación de regersión
# Iniciamos el proceso de la determinación del modelo de regresion lineal
X = pd.DataFrame({'beta0' : np.ones(df.shape[0]), 'beta1' : df ['KgInicial'], 'beta2' : df['kgAlimento'] })
#diccionario Lave : valor
# Definimos la variable de respuesta
y = df['KgFinal']
# opción para sustituir 'X'
#x_c_intercepto = sm.add_constant(X)
# Especificamos el modelo por usar
modelo = sm.OLS(y, X)
# Ajustamos el modelo
resultado = modelo.fit()
# Imprimimos los resultados
print(resultado.summary())
```
