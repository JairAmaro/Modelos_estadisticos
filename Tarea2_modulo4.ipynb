```python
import pandas as pd
import matplotlib.pyplot as plt
from io import BytesIO
import win32clipboard

# Configuración para mostrar números completos
pd.options.display.float_format = '{:.6f}'.format

# Filtrar solo columnas numéricas
df_numeric = df.select_dtypes(include=['number'])

# Obtener estadísticas descriptivas
descriptive_stats = df_numeric.describe().T
descriptive_stats = descriptive_stats[['count', 'mean', 'std', 'min']]

# Crear figura para la tabla
fig, ax = plt.subplots(figsize=(10, 6))  # Tamaño ajustable
ax.axis('tight')
ax.axis('off')
table = ax.table(cellText=descriptive_stats.values,
                 colLabels=descriptive_stats.columns,
                 rowLabels=descriptive_stats.index,
                 loc='center',
                 cellLoc='center')

# Formato del texto
table.auto_set_font_size(False)
table.set_fontsize(10)
table.auto_set_column_width(col=list(range(len(descriptive_stats.columns) + 1)))

# Guardar la imagen como bytes
buf = BytesIO()
plt.savefig(buf, format='png', bbox_inches='tight')
buf.seek(0)

# Copiar la imagen al portapapeles (Windows)
def copy_image_to_clipboard(image_data):
    win32clipboard.OpenClipboard()
    win32clipboard.EmptyClipboard()
    win32clipboard.SetClipboardData(win32clipboard.CF_DIB, image_data)
    win32clipboard.CloseClipboard()

# Convertir a formato DIB para compatibilidad
from PIL import Image
img = Image.open(buf)
output = BytesIO()
img.convert("RGB").save(output, "BMP")
data = output.getvalue()[14:]  # Eliminar cabecera de BMP
output.close()

copy_image_to_clipboard(data)
buf.close()

print("La tabla ha sido copiada como imagen al portapapeles. ¡Listo para pegar en Word!")
```

```python
# Seleccionar únicamente las columnas numéricas
numerical_columns = df.select_dtypes(include=['number'])

# Calcular la varianza para todas las columnas numéricas
variances = numerical_columns.var()

# Filtrar las variables con varianza mayor al umbral (por ejemplo, 2)
threshold = 2
high_variance_vars = variances[variances > threshold]

# Ordenar las variables por varianza de mayor a menor
high_variance_vars = high_variance_vars.sort_values(ascending=False)

# Mostrar las 5 variables con mayor varianza
top_5_vars = high_variance_vars.head(6)
print("Variables con mayor varianza:")
print(top_5_vars)
```
