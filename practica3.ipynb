Supongamos que los valores de un conjunto determinado de datos se
agrupan en intervalos. Los intervalos y las frecuencias correspondientes
son los 

Edad Frecuencia
1-5 650
5-15 900
15-20 1850
20-50 850
50-80 500
80-110 90

Calcular un valor medio aproximado
para los datos

```python
# Datos
intervalos = [(1, 5), (5, 15), (15, 20), (20, 50), (50, 80), (80, 110)]
frecuencias = [650, 900, 1850, 850, 500, 90]

# Calcular puntos medios
puntos_medios = [(a + b) / 2 for a, b in intervalos]

# Calcular x_i * f_i para cada intervalo
productos = [x * f for x, f in zip(puntos_medios, frecuencias)]

# Sumas totales
suma_frecuencias = sum(frecuencias)
suma_productos = sum(productos)

# Media
media = suma_productos / suma_frecuencias

# Imprimir resultados
print("Puntos medios:", puntos_medios)
print("Productos (x_i * f_i):", productos)
print("Suma de frecuencias:", suma_frecuencias)
print("Suma de productos:", suma_productos)
print(f"El valor medio aproximado es: {media:.3f}")
```

Supongamos que los datos para el análisis incluyen el atributo de Edad
Los valores de la edad para las tuplas de datos son (en orden creciente)
12, 15, 15, 16, 19,19, 20, 20, 20, 21, 22, 22, 25, 25, 25, 25, 30, 33, 33,
34, 35, 35, 35, 35, 36, 36, 42, 45, 46, 52, 71, 76, 78
a) ¿Cuál es la media de los datos? ¿Cuál es la mediana?
b) ¿Cuál es la moda de los datos?
c) Utilice la normalización min-max para transformar el valor 49 de la
antigüedad en el intervalo [0.0,1.0].
d) Trazar un histograma de equi-distancia de longitud 7

```python
import numpy as np
import matplotlib.pyplot as plt
from collections import Counter

# Datos
edades = [12, 15, 15, 16, 19, 19, 20, 20, 20, 21, 22, 22, 25, 25, 25, 25,
          30, 33, 33, 34, 35, 35, 35, 35, 36, 36, 42, 45, 46, 52, 71, 76, 78]

# a) Media y mediana
media = np.mean(edades)
mediana = np.median(edades)

# b) Moda
conteo = Counter(edades)
moda = [k for k, v in conteo.items() if v == max(conteo.values())]

# c) Normalización min-max para x = 49
min_edad, max_edad = min(edades), max(edades)
edad_normalizada = (49 - min_edad) / (max_edad - min_edad)

# d) Histograma de longitud 7
num_bins = 7
plt.hist(edades, bins=num_bins, edgecolor='black', alpha=0.7)
plt.title("Histograma de Edades (7 intervalos)")
plt.xlabel("Edad")
plt.ylabel("Frecuencia")
plt.show()

# Imprimir resultados
print(f"Media: {media:.2f}")
print(f"Mediana: {mediana}")
print(f"Moda: {moda}")
print(f"Valor normalizado (edad 49): {edad_normalizada:.2f}")
```

Supongamos que un hospital probó los datos de edad y grasa corporal para 18 adultos
seleccionados al azar con el siguiente resultado

Edad 22 23 27 27 39 42 47 49 50
%Grasa 9.5 26.5 7.7 18.8 31.4 25.9 27.5 27.2 31.3

Edad 52 54 54 56 57 59 59 62 66

b) Grafica el diagrama de dispersión

c) Normaliza los datos de Edad y % Grasa

d) Calcula y explica el coeficiente de correlación de Pearson
Edad 22 23 27 27 39 42 47 49 50


```python
import numpy as np
import matplotlib.pyplot as plt

# Datos
edad = [22, 23, 27, 27, 39, 42, 47, 49, 50, 52, 54, 54, 56, 57, 59, 59, 62, 66]
grasa = [9.5, 26.5, 7.7, 18.8, 31.4, 25.9, 27.5, 27.2, 31.3, 
         34.5, 42.5, 28.9, 33.5, 30.2, 34.1, 33.3, 41.4, 35.7]

# a) Media, mediana y desviación estándar
media_edad = np.mean(edad)
media_grasa = np.mean(grasa)
mediana_edad = np.median(edad)
mediana_grasa = np.median(grasa)
std_edad = np.std(edad)
std_grasa = np.std(grasa)

# b) Diagrama de dispersión
plt.scatter(edad, grasa, color='blue', alpha=0.7)
plt.title("Diagrama de dispersión (Edad vs % Grasa Corporal)")
plt.xlabel("Edad")
plt.ylabel("% Grasa Corporal")
plt.grid()
plt.show()

# c) Normalización Min-Max
edad_normalizada = [(x - min(edad)) / (max(edad) - min(edad)) for x in edad]
grasa_normalizada = [(x - min(grasa)) / (max(grasa) - min(grasa)) for x in grasa]

# d) Coeficiente de correlación de Pearson
r = np.corrcoef(edad, grasa)[0, 1]

# Imprimir resultados
print("**a) Estadísticas básicas**")
print(f"Media (Edad): {media_edad:.2f}, Media (% Grasa): {media_grasa:.2f}")
print(f"Mediana (Edad): {mediana_edad}, Mediana (% Grasa): {mediana_grasa}")
print(f"Desviación estándar (Edad): {std_edad:.2f}, Desviación estándar (% Grasa): {std_grasa:.2f}")

print("\n**c) Normalización Min-Max**")
print(f"Edad Normalizada: {edad_normalizada}")
print(f"% Grasa Normalizada: {grasa_normalizada}")

print("\n**d) Coeficiente de correlación de Pearson**")
print(f"r = {r:.2f}")
```

4. Distancias
Cálculo de distancias, atributos numéricos
Sean las tuplas: X1=(6,2) y X2=(5,9)
Encuentre la distancia utilizando:

a) Distancia máxima

b) Distancia de Manhattan

c) Distancia Euclidiana

Expresar gráficamente los resultados

```python
import numpy as np
import matplotlib.pyplot as plt

# Coordenadas de las tuplas
x1 = np.array([6, 2])
x2 = np.array([5, 9])

# a) Distancia máxima
distancia_max = np.max(np.abs(x1 - x2))

# b) Distancia de Manhattan
distancia_manhattan = np.sum(np.abs(x1 - x2))

# c) Distancia Euclidiana
distancia_euclidiana = np.sqrt(np.sum((x1 - x2)**2))

# Gráfica de la Distancia Máxima
plt.figure(figsize=(6, 6))
plt.scatter(*x1, color='red', label='X1 (6, 2)', s=100)
plt.scatter(*x2, color='blue', label='X2 (5, 9)', s=100)
plt.plot([x1[0], x2[0]], [x1[1], x2[1]], color='gray', linestyle='--', label=f'Distancia Máxima = {distancia_max}')
plt.title("Distancia Máxima")
plt.xlabel("X")
plt.ylabel("Y")
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(alpha=0.3)
plt.legend()
plt.show()

# Gráfica de la Distancia de Manhattan
plt.figure(figsize=(6, 6))
plt.scatter(*x1, color='red', label='X1 (6, 2)', s=100)
plt.scatter(*x2, color='blue', label='X2 (5, 9)', s=100)
# Líneas horizontales y verticales para la distancia de Manhattan
plt.plot([x1[0], x2[0]], [x1[1], x1[1]], color='green', linestyle='--', label=f'Horizontal: {abs(x1[0] - x2[0])}')
plt.plot([x2[0], x2[0]], [x1[1], x2[1]], color='purple', linestyle='--', label=f'Vertical: {abs(x1[1] - x2[1])}')
plt.title("Distancia de Manhattan")
plt.xlabel("X")
plt.ylabel("Y")
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(alpha=0.3)
plt.legend()
plt.show()

# Gráfica de la Distancia Euclidiana
plt.figure(figsize=(6, 6))
plt.scatter(*x1, color='red', label='X1 (6, 2)', s=100)
plt.scatter(*x2, color='blue', label='X2 (5, 9)', s=100)
plt.plot([x1[0], x2[0]], [x1[1], x2[1]], color='orange', linestyle='-', label=f'Distancia Euclidiana = {distancia_euclidiana:.2f}')
plt.title("Distancia Euclidiana")
plt.xlabel("X")
plt.ylabel("Y")
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(alpha=0.3)
plt.legend()
plt.show()

# Imprimir resultados
print(f"Distancia máxima: {distancia_max}")
print(f"Distancia de Manhattan: {distancia_manhattan}")
print(f"Distancia Euclidiana: {distancia_euclidiana:.2f}")
```

5. Distancias
Cálculo de distancias, atributos categóricos
Sea la siguiente tabla de datos

ID Color_ojos Sexo Edad
1 Café M 25
2 Negro F 23
3 Azul F 29
4 Negro F 27
5 Café M 23
6 Verde M 28
7 Negro F 26

a) Agrega tus datos y realiza lo siguiente

b) Utilizando distancia Euclidiana, encuentre la distancia entre las tuplas 2 y 6

c) Utilizando distancia Euclidiana, encuentre la distancia entre las tuplas 3 y 8

d) Realiza lo mismo con la distancia de Manhattan y Mahalanobis

e) Normaliza la edad


```python
import pandas as pd
import numpy as np
from scipy.spatial.distance import mahalanobis

# Datos
data = [
    {"ID": 1, "Color_ojos": "cafe", "Sexo": "M", "Edad": 25},
    {"ID": 2, "Color_ojos": "negro", "Sexo": "F", "Edad": 23},
    {"ID": 3, "Color_ojos": "azul", "Sexo": "F", "Edad": 29},
    {"ID": 4, "Color_ojos": "negro", "Sexo": "F", "Edad": 27},
    {"ID": 5, "Color_ojos": "cafe", "Sexo": "M", "Edad": 23},
    {"ID": 6, "Color_ojos": "verde", "Sexo": "F", "Edad": 28},
    {"ID": 7, "Color_ojos": "negro", "Sexo": "F", "Edad": 26},
    {"ID": 8, "Color_ojos": "cafe", "Sexo": "M", "Edad": 25},
]

# Asignación de valores numéricos a los atributos categóricos
color_map = {"cafe": 1, "negro": 2, "azul": 3, "verde": 4}
sexo_map = {"M": 0, "F": 1}

# Convertir datos a matriz numérica
numerical_data = [
    [color_map[entry["Color_ojos"]], sexo_map[entry["Sexo"]], entry["Edad"]]
    for entry in data
]

# Tuplas a comparar
tupla_2 = numerical_data[1]  # ID 2
tupla_6 = numerical_data[5]  # ID 6
tupla_3 = numerical_data[2]  # ID 3
tupla_8 = numerical_data[7]  # ID 8

# Calcular distancia Euclidiana
dist_euclid_2_6 = np.linalg.norm(np.array(tupla_2) - np.array(tupla_6))
dist_euclid_3_8 = np.linalg.norm(np.array(tupla_3) - np.array(tupla_8))

# Calcular distancia de Manhattan
dist_manhattan_2_6 = np.sum(np.abs(np.array(tupla_2) - np.array(tupla_6)))
dist_manhattan_3_8 = np.sum(np.abs(np.array(tupla_3) - np.array(tupla_8)))

# Calcular distancia de Mahalanobis
data_matrix = np.array(numerical_data)
inv_cov_matrix = np.linalg.inv(np.cov(data_matrix.T))
dist_mahalanobis_2_6 = mahalanobis(tupla_2, tupla_6, inv_cov_matrix)
dist_mahalanobis_3_8 = mahalanobis(tupla_3, tupla_8, inv_cov_matrix)

# Normalización Min-Max para la edad
edades = [entry["Edad"] for entry in data]
edad_min, edad_max = min(edades), max(edades)
edades_normalizadas = [(edad - edad_min) / (edad_max - edad_min) for edad in edades]

# Crear DataFrame con los resultados
df_resultados = pd.DataFrame(data)
df_resultados["Edad Normalizada"] = edades_normalizadas

# Resultados
print("Distancia Euclidiana entre tupla 2 y 6:", dist_euclid_2_6)
print("Distancia Euclidiana entre tupla 3 y 8:", dist_euclid_3_8)
print("Distancia Manhattan entre tupla 2 y 6:", dist_manhattan_2_6)
print("Distancia Manhattan entre tupla 3 y 8:", dist_manhattan_3_8)
print("Distancia Mahalanobis entre tupla 2 y 6:", dist_mahalanobis_2_6)
print("Distancia Mahalanobis entre tupla 3 y 8:", dist_mahalanobis_3_8)
print("\nTabla de datos con edad normalizada:")
print(df_resultados)
```
