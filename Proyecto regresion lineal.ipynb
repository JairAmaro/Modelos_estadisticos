# Transformaciones de Datos

Las transformaciones de datos son técnicas utilizadas para modificar la forma de los datos de modo que se adecuen mejor a los modelos estadísticos o de aprendizaje automátic

Estas transformaciones pueden ayudar a mejorar la precisión de los modelos y a cumplir con los supuestos estadísticos necesarios

A continuación, se describen varias transformaciones comunes y su utilidad

## Tipos de Transformaciones

1. **Transformación Logarítmica**
   - **Descripción**: Esta transformación aplica la función logarítmica a los datos

   - **Utilidad**: Es útil para reducir la asimetría en distribuciones sesgadas a la 
   derecha y para estabilizar la varianza. Se utiliza frecuentemente en datos que siguen una distribución exponencial

   - **Fórmula**: 
     $$
     Y' = \log(Y)
     $$
   - **Ejemplo**: Si $ Y = 1000 $, entonces $ Y' = \log(1000) \approx 6.907 $

2. **Transformación de Potencia (Box-Cox)**
   - **Descripción**: Esta transformación es un método flexible que puede aplicar varias transformaciones de potencia según el parámetro $\lambda$

   - **Utilidad**: Se utiliza para normalizar datos y estabilizar la varianza. La elección de $\lambda$ determina el tipo de transformación aplicada

   - **Fórmula**:
     $$
     Y' = 
     \begin{cases}
       \frac{Y^\lambda - 1}{\lambda} & \text{si } \lambda \neq 0 \\
       \log(Y) & \text{si } \lambda = 0
     \end{cases}
     $$

3. **Transformación de Raíz Cuadrada**
   - **Descripción**: Se aplica la raíz cuadrada a los datos

   - **Utilidad**: Ayuda a reducir la asimetría en distribuciones sesgadas a la derecha y a manejar la heterocedasticidad en los errores

   - **Fórmula**:
     $$
     Y' = \sqrt{Y}
     $$
   - **Ejemplo**: Si $ Y = 9 $, entonces $ Y' = \sqrt{9} = 3 $

4. **Transformación de Z-score (Estandarización)**
   - **Descripción**: Se centra en la media y la desviación estándar de los datos

   - **Utilidad**: Convierte los datos a una distribución con media 0 y desviación estándar 1, lo que facilita la comparación entre diferentes conjuntos de datos

   - **Fórmula**:
     $$
     Z = \frac{Y - \mu}{\sigma}
     $$
   - Donde $ \mu $ es la media y $ \sigma $ es la desviación estándar

5. **Normalización Min-Max**
   - **Descripción**: Se transforma los datos para que se encuentren dentro de un rango específico, típicamente [0, 1]

   - **Utilidad**: Es útil cuando se requiere que los datos se mantengan en un rango específico, lo cual es importante para algoritmos que son sensibles a las escalas de los datos, como las redes neuronales

   - **Fórmula**:
     $$
     Y' = \frac{Y - Y_{min}}{Y_{max} - Y_{min}}
     $$

## Conclusión

* Las transformaciones de datos son herramientas esenciales en el análisis y modelado de datos

* La elección de la transformación adecuada depende de la naturaleza de los datos y del objetivo del análisis

* Aplicar estas transformaciones puede llevar a mejoras significativas en el rendimiento de los modelos y a interpretaciones más precisas de los resultados

```python
# Mejorar error de inprenta hacer manual el OLS recordatorio *-*--*-*-*-*-*

# Codificación del género como variable binaria
df['genero_binario'] = df['genero'].apply(lambda x: 1 if x == 'F' else 0)  # 1 para F, 0 para M

# Ajuste del modelo de regresión lineal
X = df[['ingreso', 'miembros de familia', 'genero_binario']]
y = df['utilidad']
X = sm.add_constant(X)  # Añadir constante al modelo

model = sm.OLS(y, X).fit()  # Ajustar el modelo
```

```python
print(model.summary())
```

```python
# Función para calcular la regresión lineal
def linear_regression(X, y):
    X_b = np.c_[np.ones(X.shape[0]), X]            # Añadir el término independiente
    beta = np.linalg.inv(X_b.T @ X_b) @ X_b.T @ y  # Cálculo de los coeficientes
    return beta

# Seleccionar las variables independientes
X_manual = df[['ingreso', 'miembros de familia', 'genero_binario']].values
y_manual = df['utilidad'].values

# Calcular los coeficientes manualmente
coeficientes_manual = linear_regression(X_manual, y_manual)
```
